```python

import datetime

# Get current date, month, year, hour, minute, and timestamp
current_datetime = datetime.datetime.now()

current_date = current_datetime.strftime("%m/%d/%Y, %H:%M:%S")
current_day = current_datetime.day
current_month = current_datetime.month
current_year = current_datetime.year
current_hour = current_datetime.hour
current_minute = current_datetime.minute
current_timestamp = current_datetime.timestamp()

print("Current Date:", current_date)
print("Current Day:", current_day)
print("Current Month:", current_month)
print("Current Year:", current_year)
print("Current Hour:", current_hour)
print("Current Minute:", current_minute)
print("Current Timestamp:", current_timestamp)

# Change "5 December, 2019" to time
given_time_string = "5 December, 2019"
given_time = datetime.datetime.strptime(given_time_string, "%d %B, %Y").time()
print("Given Time:",given_time_string, given_time)

# Calculate time difference between now and new year
new_year = datetime.datetime(current_year + 1, 1, 1)
time_until_new_year = new_year - current_datetime
print("Time until New Year:", time_until_new_year)

# Calculate time difference between 1 January 1970 and now
epoch_time = datetime.datetime.utcfromtimestamp(0)
time_since_epoch = current_datetime - epoch_time
print("Time since 1 January 1970:", time_since_epoch)

# Formatted current date
formatted_current_date = current_datetime.strftime("%m/%d/%Y, %H:%M:%S")
print("Formatted Current Date:", formatted_current_date)
```
