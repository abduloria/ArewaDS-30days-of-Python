```python

import requests
from collections import Counter
from bs4 import BeautifulSoup

romeo_and_juliet_url = 'http://www.gutenberg.org/files/1112/1112.txt'
response = requests.get(romeo_and_juliet_url)
text = response.text

# Using BeautifulSoup to clean HTML tags if any
soup = BeautifulSoup(text, 'html.parser')
cleaned_text = soup.get_text()

# Tokenizing words and finding the 10 most frequent ones
words = cleaned_text.split()
word_counts = Counter(words)
most_frequent_words = word_counts.most_common(10)

print("10 Most Frequent Words in Romeo and Juliet:")
print(most_frequent_words)
```

```python
import requests
import statistics

cats_api_url = 'https://api.thecatapi.com/v1/breeds'
response = requests.get(cats_api_url)
cats_data = response.json()

# Extract weights as floats and lifespans as integers
weights = [float(cat['weight']['metric'].split()[0]) for cat in cats_data]
lifespans = [int(lifespan.split()[0]) if lifespan.split()[0].isdigit() else None for lifespan in lifespans]

# Analyzing weights
min_weight, max_weight, mean_weight, median_weight, std_dev_weight = (
    min(weights), max(weights), statistics.mean(weights),
    statistics.median(weights), statistics.stdev(weights)
)

# Analyzing lifespans
min_lifespan, max_lifespan, mean_lifespan, median_lifespan, std_dev_lifespan = (
    min(lifespans), max(lifespans), statistics.mean(lifespans),
    statistics.median(lifespans), statistics.stdev(lifespans)
)

print("Weight Analysis:")
print(f"Min: {min_weight}, Max: {max_weight}, Mean: {mean_weight}, Median: {median_weight}, Std Dev: {std_dev_weight}")

print("\nLifespan Analysis:")
print(f"Min: {min_lifespan}, Max: {max_lifespan}, Mean: {mean_lifespan}, Median: {median_lifespan}, Std Dev: {std_dev_lifespan}")
```

```python

import requests

countries_api_url = 'https://restcountries.eu/rest/v2/all'
countries_data = requests.get(countries_api_url, timeout=20).json()

largest_countries = sorted(countries_data, key=lambda x: x['area'], reverse=True)[:10]
print("10 Largest Countries:")
for country in largest_countries:
    print(f"{country['name']} - Area: {country['area']} sq km")

most_spoken_languages = [language['name'] for country in countries_data for language in country['languages']]
all_languages = set(most_spoken_languages)

print("\n10 Most Spoken Languages:")
for language in most_spoken_languages[:10]:
    print(language)

print(f"\nTotal Number of Languages: {len(all_languages)}")
```
