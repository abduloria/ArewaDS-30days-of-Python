```python

import random
import string

# Level 1

def random_user_id():
    # Generate a six-digit/character random_user_id
    return ''.join(random.choice(string.ascii_lowercase + string.digits) for _ in range(6))

print(random_user_id())

def user_id_gen_by_user():
    # Declare a function to generate user-defined user_ids
    num_characters, num_ids = map(int, input("Enter number of characters and number of IDs (separated by space): ").split())
    return [''.join(random.choice(string.ascii_lowercase + string.digits) for _ in range(num_characters)) for _ in range(num_ids)]

print(user_id_gen_by_user())
print(user_id_gen_by_user())

def rgb_color_gen():
    # Generate RGB colors with values ranging from 0 to 255
    return f"rgb({random.randint(0, 255)}, {random.randint(0, 255)}, {random.randint(0, 255)})"

# Level 2

def list_of_hexa_colors(num_colors):
    # Return an array of hexadecimal colors
    return ['#' + ''.join(random.choice('0123456789abcdef') for _ in range(6)) for _ in range(num_colors)]

def list_of_rgb_colors(num_colors):
    # Return an array of RGB colors
    return [rgb_color_gen() for _ in range(num_colors)]

def generate_colors(color_type, num_colors):
    # Generate hexa or rgb colors based on user input
    if color_type == 'hexa':
        return list_of_hexa_colors(num_colors)
    elif color_type == 'rgb':
        return list_of_rgb_colors(num_colors)

print(generate_colors('hexa', 3))
print(generate_colors('hexa', 1))
print(generate_colors('rgb', 3))
print(generate_colors('rgb', 1))

# Level 3

def shuffle_list(input_list):
    # Shuffle a given list
    shuffled_list = input_list.copy()
    random.shuffle(shuffled_list)
    return shuffled_list

def seven_unique_random_numbers():
    # Return an array of seven unique random numbers in the range of 0-9
    return random.sample(range(10), 7)

# Example usage:
my_list = [1, 2, 3, 4, 5, 6, 7]
print(shuffle_list(my_list))

print(seven_unique_random_numbers())
```
