```python

# Level 1

# 1. Add two numbers
def add_two_numbers(a, b):
    result = a + b
    print(f"The sum of {a} and {b} is: {result}")
    return result

# 2. Area of a circle
def area_of_circle(radius):
    area = 3.14 * radius * radius
    print(f"The area of a circle with radius {radius} is: {area}")
    return area

# 3. Add all numbers
def add_all_nums(*args):
    if all(isinstance(num, (int, float)) for num in args):
        total = sum(args)
        print(f"The sum of all numbers is: {total}")
        return total
    else:
        print("All list items should be numbers.")

# 4. Celsius to Fahrenheit
def convert_celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    print(f"{celsius}°C is equal to {fahrenheit}°F")
    return fahrenheit

# 5. Check season
def check_season(month):
    seasons = {1: 'Winter', 2: 'Winter', 3: 'Spring', 4: 'Spring', 5: 'Spring', 6: 'Summer', 7: 'Summer', 8: 'Summer', 9: 'Autumn', 10: 'Autumn', 11: 'Autumn', 12: 'Winter'}
    result = seasons.get(month, 'Invalid month')
    print(f"The season for month {month} is: {result}")
    return result

# 6. Calculate slope
def calculate_slope(a, b):
    slope = b / a
    print(f"The slope of the linear equation is: {slope}")
    return slope

# 7. Solve quadratic equation
def solve_quadratic_eqn(a, b, c):
    discriminant = b**2 - 4*a*c
    root1 = (-b + discriminant**0.5) / (2*a)
    root2 = (-b - discriminant**0.5) / (2*a)
    print(f"The solutions to the quadratic equation are: {root1}, {root2}")
    return root1, root2

# 8. Print list
def print_list(lst):
    for item in lst:
        print(item)

# 9. Reverse list
def reverse_list(arr):
    reversed_arr = arr[::-1]
    print(f"The reversed list is: {reversed_arr}")
    return reversed_arr

# 10. Capitalize list items
def capitalize_list_items(lst):
    capitalized_lst = [item.capitalize() for item in lst]
    print(f"The capitalized list is: {capitalized_lst}")
    return capitalized_lst

# 11. Add item
def add_item(lst, item):
    lst.append(item)
    print(f"The list after adding {item} is: {lst}")
    return lst

# 12. Remove item
def remove_item(lst, item):
    if item in lst:
        lst.remove(item)
    print(f"The list after removing {item} is: {lst}")
    return lst

# 13. Sum of numbers
def sum_of_numbers(n):
    total = sum(range(1, n + 1))
    print(f"The sum of numbers from 1 to {n} is: {total}")
    return total

# 14. Sum of odds
def sum_of_odds(n):
    odd_sum = sum(range(1, n + 1, 2))
    print(f"The sum of odd numbers from 1 to {n} is: {odd_sum}")
    return odd_sum

# 15. Sum of evens
def sum_of_evens(n):
    even_sum = sum(range(2, n + 1, 2))
    print(f"The sum of even numbers from 1 to {n} is: {even_sum}")
    return even_sum

# Level 2

# 1. Evens and Odds
def evens_and_odds(num):
    evens = sum(1 for digit in str(num) if int(digit) % 2 == 0)
    odds = sum(1 for digit in str(num) if int(digit) % 2 != 0)
    print(f"The number of odds are {odds}.\nThe number of evens are {evens}.")
    return odds, evens

# 2. Factorial
def factorial(n):
    if n == 0 or n == 1:
        result = 1
    else:
        result = n * factorial(n-1)
    print(f"The factorial of {n} is: {result}")
    return result

# 3. Is Empty
def is_empty(value):
    result = not bool(value)
    print(f"The input is empty: {result}")
    return result


# Level 3

# 1. Is Prime
def is_prime(num):
    if num < 2:
        result = False
    else:
        result = all(num % i != 0 for i in range(2, int(num**0.5) + 1))
    print(f"{num} is prime: {result}")
    return result

# 2. All items unique
def all_items_unique(lst):
    result = len(lst) == len(set(lst))
    print(f"All items in the list are unique: {result}")
    return result

# 3. Same data type
def same_data_type(lst):
    result = all(type(item) == type(lst[0]) for item in lst)
    print(f"All items in the list have the same data type: {result}")
    return result

# 4. Valid Python variable
def valid_python_variable(variable):
    try:
        exec(f"{variable} = 1")
        del locals()[variable]
        result = True
    except:
        result = False
    print(f"{variable} is a valid Python variable: {result}")
    return result

# 5. Most Spoken Languages
def most_spoken_languages():
    languages = ["English", "Mandarin", "Spanish", "Hindi", "Arabic", "Bengali", "Portuguese", "Russian", "Urdu", "Indonesian"]
    print(f"The most spoken languages are: {languages}")
    return languages

# 6. Most Populated Countries
def most_populated_countries():
    countries = ["China", "India", "United States", "Indonesia", "Pakistan", "Brazil", "Nigeria", "Bangladesh", "Russia", "Mexico"]
    print(f"The most populated countries are: {countries}")
    return countries
```
