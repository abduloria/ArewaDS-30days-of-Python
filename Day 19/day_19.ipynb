```python
# Level 1:Count Lines and Words

def count_lines_and_words(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
        num_lines = len(lines)
        words = [word for line in lines for word in line.split()]
        num_words = len(words)
    return num_lines, num_words

filename = '/content/drive/MyDrive/Colab Notebooks/obama_speech.txt'
lines, words = count_lines_and_words(filename)

print("Number of lines:", lines)
print("Number of words:", words)

import os

def count_lines_words(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        num_lines = len(lines)
        words = [word for line in lines for word in line.split()]
        num_words = len(words)
    return num_lines, num_words

def count_lines_words_in_files():
    speeches = [
        "obama_speech.txt",
        "michelle_obama_speech.txt",
        "donald_speech.txt",
        "melina_trump_speech.txt"
    ]

    folder_path = os.path.join(os.getcwd(), "/content/drive/MyDrive/Colab Notebooks")  # Update to the correct path

    for speech in speeches:
        file_path = os.path.join(folder_path, speech)
        if os.path.exists(file_path):
            num_lines, num_words = count_lines_words(file_path)
            print(f"File: {speech}")
            print(f"Number of lines: {num_lines}")
            print(f"Number of words: {num_words}")
            print("-" * 30)
        else:
            print(f"File {speech} not found!")

count_lines_words_in_files()
```

```python

# Level 2:

# Extract Email Addresses:
import re

def extract_emails(filename):
    with open(filename, 'r') as file:
        text = file.read()
        emails = re.findall(r'\S+@\S+', text)
    return emails

# Example usage:
print(extract_emails('/content/drive/MyDrive/Colab Notebooks/email_exchange_big.txt'))

#Most Common Words:

from collections import Counter
import re

def find_most_common_words(text, n):
    words = re.findall(r'\b\w+\b', text.lower())
    word_counts = Counter(words)
    common_words = [(count, word) for word, count in word_counts.most_common(n)]
    return common_words

# Example usage:
print(find_most_common_words('sample.txt', 10))
```
